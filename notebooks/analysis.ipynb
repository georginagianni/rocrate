```python
import pandas as pd
from pathlib import Path

# If notebook is opened from /notebooks, go up one level to crate root
if Path.cwd().name == "notebooks":
    base_dir = Path.cwd().parent
else:
    base_dir = Path.cwd()

data_path = base_dir / "data" / "speeches.csv"
output_dir = base_dir / "outputs"
output_file = output_dir / "word_counts.csv"

output_dir.mkdir(parents=True, exist_ok=True)

data = pd.read_csv(data_path)
word_counts = data["text"].str.split().explode().value_counts()
word_counts.to_csv(output_file)

print("Notebook analysis complete")
print(f"Output written to: {output_file}")
```

## Interpretation for SSH Research

The FLAT tool is integrated into the research environment as a machine-readable resource using CodeMeta metadata. 

Instead of manually searching for tools, researchers can:
- discover tools through structured metadata
- understand their purpose and requirements
- reuse them within reproducible workflows

This demonstrates how external SSH tools can be incorporated into executable research environments without requiring local installation.
